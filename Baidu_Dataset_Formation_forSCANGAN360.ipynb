```python
import os
import posixpath
import pandas as pd
import time
```

```python
main_gaze_df = pd.DataFrame(columns = ["Video", "Participant", "Coordinates"])
main_head_df = pd.DataFrame(columns = ["Video", "Participant", "Coordinates"])

path = 'C:/Users/arshk/MLAI/Research Project/Baidu/Gaze_txt_files'
cols = ['frame', 'frame index', 'forward', 'head position x', 'head position y', 'eye', 'gaze position x', 'gaze position y']

start_time = time.time()

for i in os.listdir(path):
    x = i.split()
    participant_no = x[0]
    print(participant_no)
    p_time = start_time
    if x[0][0] == 'p':
        new_path = posixpath.join(path, i, "")
        p_time = time.time()
        for j in os.listdir(new_path):
            file_path = posixpath.join(new_path, j)
            n = j.split()
            video_no = n[0][:3]
            
            temp_df = pd.read_csv(file_path, sep=",", header = None, names = cols)
                            
            list_gaze = []
            list_head = []
            
            for k in range(0, len(temp_df)):
                list_gaze.append(temp_df.loc[k]["gaze position x"])
                list_gaze.append(temp_df.loc[k]["gaze position y"])
                
                list_head.append(temp_df.loc[k]["head position x"])
                list_head.append(temp_df.loc[k]["head position y"])
            
            dict_gaze = {'Video': video_no, 'Participant': participant_no, 'Coordinates': list_gaze}
            dict_head = {'Video': video_no, 'Participant': participant_no, 'Coordinates': list_head}
            main_gaze_df = main_gaze_df.append(dict_gaze, ignore_index = True)
            main_head_df = main_head_df.append(dict_head, ignore_index = True)
    print("Length of main_gaze_df = " + str(len(main_gaze_df)) + ". Length of main_head_df = " + str(len(main_head_df)))
    print("Time for participant " + str(participant_no) + " = " + str(time.time()-p_time))

print("Time elapsed = " + str(time.time()))
```

```python
main_gaze_df.sort_values(["Video", "Participant"], inplace = True)
```

```python
main_head_df.sort_values(["Video", "Participant"], inplace = True)
```

```python
main_gaze_df.to_csv("Gazes.csv", index = False)
main_head_df.to_csv("Heads.csv", index = False)
```
